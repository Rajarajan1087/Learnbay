```python
# leap year - Any year(exception hundred year e.g 1600, 1800, 1900) divisible by 4 is a leap year
# if year is hundred year(00) -> it shold be divisible by 400. then it will be leap year

# 1997
# 2004
# 1800
# 1600

year = int(input('Enter an year: '))
if year%4 == 0:
  if year%100 == 0:
    if year%400 == 0:
      print(f'{year} is a leap year')
    else:
      print(f'{year} is not a leap year')
  else:
    print(f'{year} is a leap year')
else:
  print(f'{year} is not a leap year')
```
