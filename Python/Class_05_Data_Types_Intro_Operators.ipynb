```python
# Mutable - Modify - Item assignment is possible - id will not change
lst1 = [1, 2, 3, 4]
print('Before Operation')
print(lst1, id(lst1))
# Accessing the element/item in list/sequence - indexing
# print(lst1[1])
# # Accessing last element in the list
# print(lst1[-1])
print()
print('After Operation')
lst1[1] = 20
print(lst1, id(lst1))

```

```python
# Arithmetic -> +, -, *, /(Float Division), //(Floor Division), %(Modulo), **(Exponent)

a = 100
b = 3
# print(a+b)
# print(a-b)
# print(a*b)

# floor and ceil

# ceil - higest nearest whole number - ceil
# floor - lowet nearest whole number - floor

# ceil - 11, 11, -4
# 10.5, 10.9, -4.5
# floor - 10, 10, -5

import math
print(f"floor of 10.5 -> {math.floor(10.5)}")
print(f"ceil of 10.5 -> {math.ceil(10.5)}")

print(f"floor of 10.9 -> {math.floor(10.9)}")
print(f"ceil of 10.9 -> {math.ceil(10.9)}")

print(f"floor of -4.9 -> {math.floor(-4.9)}")
print(f"ceil of -4.9 -> {math.ceil(-4.9)}")


# # float division - after dividing it returns a float value
print(math.ceil(a/b))
# print(100/2)

# floor division - returns floor value of float division result
print(a//b)
# same as below
print(math.floor(a/b))

# Modulo operation - % - returns remainder after division
# print(a%b)

# Exponent - a**b -> a raise to the power b
print(100**3)




```
